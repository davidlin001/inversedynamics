Training a inverse dynamics model with a FC net. Sample state-action pairs are generated with a random exploration policy for Fetch-Push. Currently using CPU

To speed up data generation process, make sure to create the initial the data matrix, and modify its elements. As it is right now, new samples are being appended to the data matrix (slow for large N).

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.utils.data import sampler
from torch.utils.data import TensorDataset

import torchvision.datasets as dset
import torchvision.transforms as T

import numpy as np

import gym 
```

```python
num_episodes = 2000
# Note that many tasks already have max episode length implicitlty defined
max_steps = 50

env = gym.make('FetchPush-v1')
state_inputs = np.zeros((0,50))
action_outputs = np.zeros((0,4))

for iepisode in range(num_episodes):
    observation = env.reset()
    for t in range(max_steps):
        # env.render()
        
        # Take random step 
        obs_before =  np.copy(observation['observation'])
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        obs_after = np.copy(observation['observation'])
        
        # Extracting input/ output info 
        state_info = np.hstack((obs_before, obs_after))
        state_inputs = np.vstack((state_inputs, state_info))
        action_outputs = np.vstack((action_outputs, action))
        if done:
            # print("finished after {} timesteps".format(t+1))
            print("Done with ", iepisode)
            break

# Split into train/ validation. Convert to tensor
train_len = (int)(np.shape(state_inputs)[0]*.8)
val_len = (np.shape(state_inputs)[0] - train_len
train_input = state_inputs[:train_len,:]
train_output = action_outputs[:train_len,:]
val_input = state_inputs[train_len:,:]
val_output = action_outputs[train_len:, :]

# Convert np array to torch tensor
train_input = torch.from_numpy(train_input)
train_output = torch.from_numpy(train_output)
val_input = torch.from_numpy(val_input)
val_output = torch.from_numpy(val_output)

# Training and Validation Sets 
train_dataset = TensorDataset(train_input, train_output)
train_dataloader = DataLoader(train_dataset, batch_size = 64, sampler = sampler.SubsetRandomSampler(range(train_len)))         
val_dataset = TensorDataset(val_input, val_output)
val_dataloader = DataLoader(val_dataset, batch_size = 64, sampler = sampler.SubsetRandomSampler(range(val_len)))    
```

```python
val_dataset = TensorDataset(val_input, val_output)
val_dataloader = DataLoader(val_dataset, batch_size = 64, sampler = sampler.SubsetRandomSampler(range(val_len)))
```

```python
class ThreeLayerNet(torch.nn.Module):
    def __init__(self, D_in, H1, H2, D_out):
        super(ThreeLayerNet, self).__init__()
        self.linear1 = nn.Linear(D_in, H1)
        self.linear2 = nn.Linear(H1, H2)
        self.linear3 = nn.Linear(H2, D_out)

    def forward(self, x):
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        return self.linear3(x)
```

```python
device = torch.device('cpu')
dtype = torch.float32
# device = torch.device('cuda') # uncomment for gpu

N, D_in, H1, H2, D_out = 64, 50, 256, 256, 4
model = ThreeLayerNet(D_in, H1, H2, D_out)
loss_fn = torch.nn.MSELoss(size_average = False)

learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)

for t, (x, y) in enumerate(train_dataloader):
    x = x.to(device=device, dtype=dtype)
    y = y.to(device=device, dtype=dtype)
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
   
for t, (x,y) in enumerate(val_dataloader):
    x = x.to(device=device, dtype=dtype)
    y = y.to(device=device, dtype=dtype)
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.item())

```
